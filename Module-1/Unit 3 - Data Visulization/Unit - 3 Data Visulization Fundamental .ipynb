```python
import matplotlib.pyplot as plt

x = [1,2,3]

y = [2,4,1]

plt.plot(x,y)

plt.xlabel('X - axis')
plt.ylabel('Y - axis')

plt.title('My First Graph')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0,10, 0.1)
y=np.sin(x)

plt.figure(figsize=(10,8))
plt.plot(x,y)
plt.show()
```
